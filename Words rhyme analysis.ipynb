```python
vowels = {"а", "ъ", "о", "у", "е", "и", "ю", "я"}
def get_vowels(word):
    build = ""
    for char in word:
        if char in vowels:
            build += char
    
    return build
```

```python
vwords = list(map(get_vowels, words))
vwords
```

```python
def one_endswith_the_other(word1, word2):
    return word1.endswith(word2) or word2.endswith(word1)

def get_vowel_rhymes(of_word, words):
    vword = get_vowels(of_word)
    return [word for word in words 
            if one_endswith_the_other(get_vowels(word), vword) and word != of_word]
```

```python
get_vowel_rhymes('автомобил', words)
```

```python
def get_last_syllable(of_word):
    syllable = ""
    cnt_vowels = 0
    
    cur_vowels = get_vowels(of_word)
    if len(cur_vowels) == 1:
        return of_word
    
    first_consonants = 0
    
    for char in reversed(of_word):
        if char in vowels:
            cnt_vowels += 1
            if cnt_vowels > 1:
                break
        
        syllable += char
        if cnt_vowels > 0:
            first_consonants += 1
    
    to_start = first_consonants // 2
    return syllable[to_start::-1]
    
get_last_syllable('ръжда')
```

```python
def get_last_syllable_rhymes(of_word, words):
    last_syllable = get_last_syllable(of_word)
    
    return [word for word in words if word.endswith(last_syllable) and word != of_word]
```

```python
get_last_syllable_rhymes('костен', words)
```

```python
def compose(of_word, with_words, rhyme_schemes):
    rhymes = with_words
    for scheme in rhyme_schemes:
        rhymes = scheme(of_word, rhymes)
    return rhymes

compose('автомобил', words, [get_vowel_rhymes, get_last_syllable_rhymes])
```
